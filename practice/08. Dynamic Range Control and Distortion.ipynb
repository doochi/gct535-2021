```python
import librosa
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
import pytsmod as tsm
```

# Distortion

```python
x, sr = librosa.load("../audio/16054__lg__guitar-clean-rif.wav") 

g = 100 # gain
x2 = g*x

y = np.zeros(len(x))

for i in range(len(x)):
    if x2[i] < -1:
        y[i] = -2/3
    elif x2[i] > 1:
        y[i] = 2/3
    else:
        y[i] = x2[i] - x2[i]*x2[i]*x2[i]/3

ipd.Audio(y, rate=sr)
```
